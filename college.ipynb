For my data exploration I rather than looking for anything, mainly want to pursue a few specific questions.
* Does performance correlate with more expenditure? What about staff variables?
* Are people in elite schools or top performers rich kids? Can only be estimated.

I estimate perfomance by looking at graduation variable, since I don't really have anything else indicating performance. Will compare it with expenditure and staff using scatter plot.

```python
pd.plotting.scatter_matrix(college[["Grad.Rate", "Expend", "S.F.Ratio", "PhD", "Personal"]], figsize=(10, 10))
```

Only interested in comparing each to graduation rate. Using my human intution to find relationships.
* Compared to expenditure. It seems like there is some postitive relationship, however it is not linear. More like it caps out at about 30k. Also there is a bit of a lack of data for higher expenditure so perhaps not very statistically reliable without more samples.
* Compared to student faculty ratio. I would say it seems uncorrelated.
* Compared with percent of faculty with phd. It seems to me like there is a bit of positive correlation.
* Compared with personal spending. It seems to be uncorrelated. Although if you go higher personal expenditure, the average does go up after a threshold. However there are not many data samples for high personal spending.



Now I want to investigate if people in elite schools are rich kids. Rich kid is estimated from the variable of personal spending, which I assume will go up if richer.

```python
college.boxplot(column="Personal", by="Elite")
```

It doesn't seem like being in an elite school entails higher personal spending. In fact it seems the median is slightly lower for elite schools with a more narrow spread. So assuming that being rich means more personal spending, then there is no indication that being wealthy is important to get into an elite university.

Summary data investigation:
* I compared some variables to graduation rate. Only instructional expenditure and percentage of PhD in faculty seem to show some kind of positive relationship. The other variables seemed uncorrelated.
* If assume that rich pepole have higher personal spending, then it doesn't seem like wealth is required to get into elite schools.
